## Intro

```python
a = np.array([[0,0,0],
                [1,1,1]])
print(a)
# 2 axes, the first axis has a length of 2 and the second axis has a length of 3.
```

`np.eye(n,m)` defines a 2D identity matrix.

`np.diag` can define either a square 2D array with given values along the diagonal or if given a 2d array returns a 1D array that is only the diagonal elements.


```python
print(np.eye(3))

print(np.eye(3,5))
```

```python
print(np.diag([1,2,3]))
```

```python
print(np.diag([1,2,3], 1))
```

```python
a = np.array([[1,2],
              [3,4]])
print(np.diag(a))
```

### Adding, Removing, and Sorting Elements
`np.sort()` and `np.concatenate()`

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
print(np.concatenate((a,b)))
```

```python
a = np.array([[1,2],
              [3,4]])
b = np.array([[5,6]])
print(np.concatenate((a,b),axis=0))
```

```python
a = np.array([1,2,3,4,5,6])
print(a.shape)
a2 = a[np.newaxis, :]
print(a2)
print(a2.shape)
```

```python
row_vector = a[np.newaxis, :]
col_vector = a[:, np.newaxis]
print(row_vector.shape)
print(col_vector.shape)
print(col_vector)
```

```python
# you can also expand an array by inserting a new axis at a pecified position 
a3 = np.expand_dims(a, axis = 0)
a4 = np.expand_dims(a, axis=1)
print(a3.shape)
print(a4.shape)
```

### Indexing and Slicing

```python
a = np.arange(12).reshape((3,4))
print(a)
```

```python
print(a[a<5])
```

```python
five_up = (a>=5)
print(a[five_up])
```

```python
c = a[(a>2) & (a<11)]
print(c)
```

```python
five_up = (a>5) | (a==5)
print(five_up)
```
